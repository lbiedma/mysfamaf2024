a) $P(X ≥ 1) = P(W_1 + W_2 ≥ 1) P(U ≤ 1/2) + P(W_1 + W_2 + W_3 ≥ 1) P(U > 1/2)$

$f_{X+Y}(t) = ∫_{-∞}^{∞}f_X(x)f_Y(t-x)dx =$

$f_X(x) = I_{(0,1)}(x) = 1 si x ∈ (0,1), 0 c.c.$
$f_X f_Y = 0$ si $f_X = 0$ o $f_Y = 0$ sii $x ∉ (0, 1)$ o $x ∉ (t - 1, t)$
- Si $t < 0$, entonces $f_X f_Y = 0$
- Si $t > 2$, entonces $f_X f_Y = 0$
- Si $0 ≤ t ≤ 1$ la longitud de $(0, 1) ∩ (t-1, t) = (0, t)$ entonces $f_X f_Y = I_{(0, t)}$, luego $∫_{-∞}^{∞}f_X(x)f_Y(t-x)dx = ∫_{-∞}^{∞} I_{(0, t)} = t$
- Si $1 < t ≤ 2$ la longitud de $(0, 1) ∩ (t-1, t) = (t-1, 1)$ entonces $f_X f_Y = I_{(t-1, 1)}$, luego $∫_{-∞}^{∞}f_X(x)f_Y(t-x)dx = ∫_{-∞}^{∞} I_{(t-1, 1)} = 2-t$

Finalmente $P(W_1 + W_2 >= 1) = 1 - ∫_{-∞}^{1}f_{X+Y}(x) dx = 1 - \int_{0}^{1}x dx = 1 - \frac{x^2}{2}|_{0}^{1} = 1 - 1/2 = 1/2$


$P(W_1 + W_2 + W_3) ≥ 1$

$P(W_1 + W_2 + W_3) = P((W_1 + W_2) + W_3) \ge 1$

$f_{(W_1 + W_2) + W_3} (t) = \int_{-∞}^{∞} f_{W_1+W_2}(x) f_{W_3}(t-x) dx$

- Si $0 < x <2$, $f_{W_1+W_2} \neq 0$

Ahora veamos t
- Si $t < 0$, cuando $0 < x < 2$, $t - x < 0$, entonces $f_{W_1 + W_2} f_{W_3} = 0$
- Si $t > 3$, cuando $0 < x < 2$, $t - x > 3 - 2 = 1$, entonces $f_{W_1 + W_2} f_{W_3} = 0$
- Si $0 ≤ t ≤ 2$,
    - $t ∈ [0, 1)$: $[t-1, t) ∩ (0, 2) = (0, t)$, entonces $\int_{-∞}^{∞} f_{W_1+W_2}(x) f_{W_3}(t-x) dx = \int_0^t x dx = \frac{x^2}{2} |_0^t = \frac{t^2}{2}$
    - $t ∈ [1, 2]$: $[t-1, t) ∩ (0, 2) = [t-1, t)$, entonces $\int_{-∞}^{∞} f_{W_1+W_2}(x) f_{W_3}(t-x) dx = \int_{t-1}^{t} 2-x dx = 2x - \frac{x^2}{2} | _{t-1}^t = 2t - \frac{t^2}{2} - (2 (t-1) - \frac{(t-1)^2}{2}) = 5/2 - t$

$P(W_1 + W_2 + W_3 ≥ 1) = 1 - ∫_{-∞}^{1} f_{W_1+W_2+W_3}(t)dt = 1 - \int_{-∞}^{1}∫_{-∞}^{∞}f_{(W_1+W_2)}(x) f_{W_3}(t-x)dx dt = 1 - \int_0^1 \frac{t^2}{2}dt = 1 - \frac{1}{6} = 5/6$

Finalmente $P(X \ge 1) = P(W_1 + W_2 ≥ 1) P(U ≤ 1/2) + P(W_1 + W_2 + W_3 ≥ 1) P(U > 1/2) = \frac{1}{2}\frac{1}{2} + \frac{5}{6}\frac{1}{2} = \frac{1}{4} + \frac{5}{12} = \frac{8}{12} = \frac{2}{3}$

```python
import random

def simulacion():
    U = random.random()
    if U < 0.5:
        X = random.random() + random.random()
    else:
        X = random.random() + random.random() + random.random()

    return X

experimentos = 100000
positivos = 0
for _ in range(experimentos):
    X = simulacion()
    if X >= 1:
        positivos += 1

print(positivos / experimentos)
```
